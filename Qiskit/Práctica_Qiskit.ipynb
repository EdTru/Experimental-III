Importamos las librerias


```python
from qiskit import QuantumCircuit
from qiskit._accelerate import results
from qiskit_aer import AerSimulator
import numpy as np
import matplotlib

pi = np.pi
```

```python
theta = [0, pi/3, pi/2, pi]
shots = 100

# Creamos un simulador cu´antico
sim = AerSimulator()
# Creamos un circuito con 1 qubit y 1 bit cl´asico
qc = QuantumCircuit(1, 1)
# Aplicamos una rotaci´on Ry(theta) al qubit
qc.ry(theta[1], 0)
# Medimos el qubit en la base computacional (Z)
qc.measure(0, 0)
# Ejecutamos el circuito un n´umero finito de veces (shots)
job = sim.run(qc, shots=shots)
# Obtenemos el resultado de la simulaci´on
result = job.result()
# Extraemos los resultados de la medida
counts = result.get_counts(qc)

print(counts)

qc.draw()
```

```python
shots = 100

sim = AerSimulator()
qc = QuantumCircuit(2, 1)

qc.h(0)
qc.cx(0,1)

qc.measure_all()

job = sim.run(qc, shots=shots)

result = job.result()
counts = result.get_counts(qc)

print(counts)

qc.draw(output="mpl")
```

```python
shots = 100

sim = AerSimulator()
qc = QuantumCircuit(3, 0)

qc.h(1)
qc.cx(1,2)
qc.cx(0,1)
qc.h(0)
qc.x(2)
qc.z(2)

qc.measure_all()

job = sim.run(qc, shots=shots)

result = job.result()
counts = result.get_counts(qc)

print(counts)

qc.draw(output="mpl")
```
